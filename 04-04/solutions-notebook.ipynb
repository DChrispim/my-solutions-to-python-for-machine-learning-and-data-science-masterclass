## Question 1
### (a)
Use the [Massâ€“energy equivalence](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) from Einstein, namely

$$E=mc^2,$$

and build the necessary data to plot the relation between mass and energy. Use 11 evenly spaced values from 0 grams to 10 grams as the mass points.

```python
# Imports
from scipy.constants import c
import numpy as np
```

```python
# Build mass vector
m = np.arange(0, 11, 1)

# Compute energy from mass vector
E = m * c ** 2
```

### (b)
Use Matplotlib to plot the previous relation

```python
# Imports
import matplotlib.pyplot as plt
```

```python
# Create plot
plt.plot(m, E, linewidth=3, color="red")

# Set labels
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')

# Set title
plt.title('E=mc^2')

# Set Lower Limit and Upper Limit on x-axis
plt.xlim(0, 10)

# Show plot
plt.show()
```

___
### (c)

Plot the same data using a logarithmic scale on the y axis.

```python
# Create plot canvas
fig, ax = plt.subplots(1, 1)

# Create plot
ax.plot(m, E, linewidth=3, color="red")

# Set axe labels
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')

# Set title
ax.set_title('E=mc^2')

# Set Lower Limit and Upper Limit on x-axis
ax.set_xlim(0, 10)

# Set log scale on y-axis
ax.set_yscale("log")


# Add grid
ax.grid(which='both', axis='y')

# Show plot
plt.show()
```

## Question 2
The data was obtained from [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield) and represents the yield curve from July-16-2007 and July-16-2020. The data shows the interest paid for a US Treasury bond for a certain contract length. The labels list shows the corresponding contract length per index position.

```python
# Data

# Labels
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

# Data points
july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

### (a)
Plot both curves on the same plot. Use the labels informed.

```python
# Create plot canvas
fig, ax = plt.subplots(figsize=(8, 4))

# Create tick list
x_ticks_list = np.arange(1, len(labels)+1)


# Plot curves
ax.plot(x_ticks_list, july16_2007, lw=2, label="july16_2007")
ax.plot(x_ticks_list, july16_2020, lw=2, label="july16_2020")

# Set x ticks
ax.set_xticks(x_ticks_list)
ax.set_xticklabels(labels, fontsize=10)

# Set y limit
ax.set_ylim([-0.5, 5.5])

# Add legends
ax.legend(loc=(0.7, 0.5))

# Show plot
plt.show()
```

### (b)
Change the previous plot, move the legend outside the main Figure plot.

```python
# Create plot canvas
fig, ax = plt.subplots(figsize=(8, 4))

# Create tick list
x_ticks_list = np.arange(1, len(labels)+1)


# Plot curves
ax.plot(x_ticks_list, july16_2007, lw=2, label="july16_2007")
ax.plot(x_ticks_list, july16_2020, lw=2, label="july16_2020")

# Set x ticks
ax.set_xticks(x_ticks_list)
ax.set_xticklabels(labels, fontsize=10)

# Set y limit
ax.set_ylim([-0.5, 5.5])

# Add legends
ax.legend(loc=(1.02, 0.5))

# Show plot
plt.show()
```

### (c)
Split the plot into two subplots, one for each year.

```python
# Create subplots canvas
fig, axes = plt.subplots(2, 1, figsize=(12, 6))

# Create tick list
x_ticks_list = np.arange(1, len(labels)+1)


# Plot the first curve
axes[0].plot(x_ticks_list, july16_2007, lw=2)
axes[0].set_xticks(x_ticks_list)
axes[0].set_xticklabels(labels, fontsize=10)
axes[0].set_title("July 16th, 2007")

# Plot the second curve
axes[1].plot(x_ticks_list, july16_2020, lw=2)
axes[1].set_xticks(x_ticks_list)
axes[1].set_xticklabels(labels, fontsize=10)
axes[1].set_title("July 16th, 2020")

# Set x ticks
ax.set_xticks(x_ticks_list)
ax.set_xticklabels(labels, fontsize=10)

# Set y limit
ax.set_ylim([-0.5, 5.5])

# Add legends
ax.legend(loc=(1.02, 0.5))

# Better adjust the image inside the figure
plt.tight_layout()

# Show plot
plt.show()
```

### (d)
Change the plot to use two distinct y axis one for each year (this is not an usual plot).

```python
# Create subplots canvas
fig, axe = plt.subplots(figsize=(12, 6))

# Create tick list
x_ticks_list = np.arange(1, len(labels)+1)

# Plot the first curve
axe.plot(x_ticks_list, july16_2007, lw=2, color="blue")
axe.set_ylabel("2007", fontsize=18, color="blue")
axe.tick_params(labelsize=18)
axe.spines['left'].set_color('blue')
axe.spines['left'].set_linewidth(4)
for label in axe.get_yticklabels():
    label.set_color("blue")

# Create twin axe
axe_twin = axe.twinx()

# Plot the second curve
axe_twin.plot(x_ticks_list, july16_2020, lw=2, color="red")
axe_twin.set_ylabel("2020", fontsize=18, color="red")
axe_twin.tick_params(labelsize=18)
axe_twin.spines['right'].set_color('red')
axe_twin.spines['right'].set_linewidth(4)
for label in axe_twin.get_yticklabels():
    label.set_color("red")


# Set x ticks
axe.set_xticks(x_ticks_list)
axe.set_xticklabels(labels, fontsize=10)

# Set y limit
axe.set_ylim([-0.5, 5.5])

# Add title
axe.set_title("July 16th Yield Curves")

# Better adjust the image inside the figure
plt.tight_layout()

# Show plot
plt.show()
```
